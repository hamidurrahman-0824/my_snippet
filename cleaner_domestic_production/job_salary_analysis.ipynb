```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"C:\Users\user\my_snippet\DataSets\job_salary_prediction_dataset.csv")
```

```python
# data cleaning

nulls = df.isnull().sum()
print(f"{nulls}")

df = df.dropna(subset='job_title')
duplicates = df.duplicated().sum()
print(f"Duplicate {duplicates}")
df = df.drop_duplicates()
```

```python
# salary range
df.boxplot(by='job_title',column='salary',figsize=(10,8))
plt.xticks(rotation=45)
plt.xlabel('Job Titles')
plt.ylabel('Salary ($)')
plt.title('Salary Distribution by Job Title')
plt.suptitle('')
plt.tight_layout()
```

```python
# Skill count Distribution
df['skills_count'].plot(kind='hist',bins=20)
plt.xlabel('Skills Count')
plt.title('Skills Count Distribution')
```

```python
highest_average_salary = (df.groupby('job_title')['salary']\
    .mean()\
    .sort_values(ascending=False)
)

highest_average_salary.plot(kind='bar',colormap='Set2',stacked=False)

plt.ylabel('Average Salary')
plt.title('Average Salary by Job Title')
```

```python
# relation between job and experience

correlation = df['experience_years'].corr(df['salary'])
print("Correlation:", correlation)
# visualization
sns.regplot(
    x='experience_years',
    y='salary',
    data=df,
    scatter_kws={'alpha':0.3}
)
```

```python
# which industries pay the most
most_payer = df.groupby('industry')['salary'].agg(['mean','median'])\
    .sort_values('mean')
print(most_payer)
```

```python
# which country offer highest salaries
countries = [      'India',   'Australia',   'Singapore',      'Canada',      'Sweden',
         'USA', 'Netherlands',     'Germany',          'UK']
df[df['location'].isin(countries)].\
    groupby('location')['salary'].\
    mean().\
    sort_values(ascending=False)
```

```python
# relation between certifications and salary
df.groupby('certifications')['salary'].median().plot(kind='bar')
plt.ylabel('Median Salary')
plt.title('Certifications vs Salary')
```

```python
# do flexibility reduce salary
remote_salary = (df.groupby(['remote_work'])['salary']\
    .mean()\
    .sort_values(ascending=False))
print(remote_salary)
```

```python
# company size and salary size

table = df.groupby('company_size')['salary'].mean().sort_values(ascending=False)
```

```python
# which company size offer more flexibility

table = df.groupby('company_size').agg({
    'remote_work':'value_counts'
})
table.unstack().plot(kind='bar',colormap='Set2')

plt.xlabel('Company Size')
plt.ylabel('Count')
plt.title('Remote Work Flexibility by Company Size')
plt.xticks(rotation=0)
```
