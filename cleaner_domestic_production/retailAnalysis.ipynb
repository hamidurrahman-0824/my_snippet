```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'C:\Users\user\my_snippet\DataSets\synthetic_dataset.csv',encoding='UTF-8')
df.head()
```

```python
# Feature Engineering


df.isnull().sum()
df = df.dropna(subset='Price')
df = df.drop_duplicates()
df['Category'] = df['Category'].fillna('X')
df['Rating'] = df['Rating'].fillna(df['Rating'].median())
```

```python
# price distribution

df['Price'].plot(kind='hist',bins=30)
```

```python
# price and rating indirect connection

df.plot(kind='scatter',x='Price',y='Rating')
```

```python
average_price_rating = df.groupby('Category').agg({'Price':'mean',
                                                   'Rating':'mean'})
```

```python
# do out of stock product has higher rating
df.groupby('Stock')['Rating'].mean()
#widest price range
df.boxplot(by='Category',column='Price')
```
