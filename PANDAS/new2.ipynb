```python
import pandas as pd
df = pd.read_csv("C:\\Users\\user\\my_snippet\\coffee_shop.csv",index_col='transaction_id')
df.columns
df.head()
```

```python
df["revenue"] = df["transaction_qty"] * df["unit_price"]
daily_revenue = df.groupby("transaction_date")["revenue"].sum()  # aggregate first

import matplotlib.pyplot as plt

# Option 1: Distribution of daily revenue totals (histogram)
plt.hist(daily_revenue, bins=20)
plt.xlabel("Daily Revenue")
plt.ylabel("Frequency")
plt.title("Distribution of Daily Revenue")
plt.show()

# Option 2: Revenue over time (line plot) — probably more useful
daily_revenue.plot()
plt.xticks(rotation=90)
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Revenue Over Time")
plt.show()
```
