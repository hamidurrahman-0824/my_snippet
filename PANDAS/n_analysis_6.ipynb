```python
#can premium ticket and age make you survive an accident
premium_ticket = df['Fare'].quantile(.75)
new_df = df[(df['Fare']>premium_ticket) & (df['Age']>df['Age'].quantile(.75))]
gg = new_df.groupby('Sex')['Survived'].sum()
normal_ticket = df['Fare'].quantile(.50)
new_df2 = df[(df['Fare']<normal_ticket)]
g = new_df2.groupby('Sex')['Survived'].sum()
print(f"Total premium survivor{gg},\
      Total normal survivor {g}.")
```
